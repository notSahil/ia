```python
import requests
import yfinance as yf
import pandas as pd
from newsapi import NewsApiClient
import tweepy

# Fetch data from World Bank API
def fetch_world_bank_data(indicator, country='IN', start_year='2000', end_year='2023'):
    url = f'http://api.worldbank.org/v2/country/{country}/indicator/{indicator}'
    params = {'format': 'json', 'date': f'{start_year}:{end_year}'}
    response = requests.get(url, params=params)
    data = response.json()
    return data

# Process World Bank data
def process_world_bank_data(data):
    if len(data) > 1:
        df = pd.DataFrame(data[1])
        df = df[['date', 'value']]
        df.columns = ['Year', 'Value']
        df = df.dropna().sort_values(by='Year')
        df['Year'] = df['Year'].astype(int)
        return df
    else:
        return pd.DataFrame()

# Fetch financial data from yfinance
def fetch_financial_data(ticker):
    company = yf.Ticker(ticker)
    income_statement = company.financials.T
    balance_sheet = company.balance_sheet.T
    cash_flow = company.cashflow.T

    def preprocess_df(df):
        df.index = pd.to_datetime(df.index)
        df = df.fillna(0)
        return df

    income_statement = preprocess_df(income_statement)
    balance_sheet = preprocess_df(balance_sheet)
    cash_flow = preprocess_df(cash_flow)

    return income_statement, balance_sheet, cash_flow

# Fetch key ratios from yfinance
def fetch_key_ratios(ticker):
    company = yf.Ticker(ticker)
    info = company.info

    pe_ratio = info.get('trailingPE', None)
    debt_to_equity = info.get('debtToEquity', None)
    roe = info.get('returnOnEquity', None)

    key_ratios = {
        'PE Ratio': pe_ratio if pe_ratio is not None else 0,
        'Debt-to-Equity Ratio': debt_to_equity if debt_to_equity is not None else 0,
        'Return on Equity (ROE)': roe if roe is not None else 0
    }

    return key_ratios

# Fetch sentiment analysis from NewsAPI
def fetch_news_sentiment(api_key, query='stock market', language='en'):
    newsapi = NewsApiClient(api_key=api_key)
    articles = newsapi.get_everything(q=query, language=language)
    return articles

# Fetch sentiment analysis from Twitter
def fetch_twitter_sentiment(api_key, api_secret_key, access_token, access_token_secret, query='#stockmarket'):
    auth = tweepy.OAuthHandler(api_key, api_secret_key)
    auth.set_access_token(access_token, access_token_secret)
    api = tweepy.API(auth)
    tweets = api.search_tweets(q=query, lang='en', count=100)
    return tweets

# Fetch and process World Bank data
gdp_data = fetch_world_bank_data('NY.GDP.MKTP.KD.ZG')
gdp_df = process_world_bank_data(gdp_data)

inflation_data = fetch_world_bank_data('FP.CPI.TOTL')
inflation_df = process_world_bank_data(inflation_data)

interest_rate_data = fetch_world_bank_data('FR.INR.LEND')
interest_rate_df = process_world_bank_data(interest_rate_data)

combined_df = gdp_df.merge(inflation_df, on='Year', suffixes=('_GDP', '_Inflation'))
combined_df = combined_df.merge(interest_rate_df, on='Year')
combined_df.columns = ['Year', 'GDP Growth Rate (%)', 'Inflation Rate (%)', 'Interest Rate (%)']

print("Combined Economic Data:")
print(combined_df.head())

# Fetch financial data and key ratios for a specific ticker
ticker = 'RELIANCE.NS'
income_statement, balance_sheet, cash_flow = fetch_financial_data(ticker)
key_ratios = fetch_key_ratios(ticker)

print("\nCleaned Income Statement:")
print(income_statement.head())

print("\nCleaned Balance Sheet:")
print(balance_sheet.head())

print("\nCleaned Cash Flow Statement:")
print(cash_flow.head())

print("\nKey Ratios:")
print(key_ratios)

# Fetch news sentiment (replace 'your_newsapi_key' with your actual NewsAPI key)
news_api_key = 'your_newsapi_key'
news_sentiment = fetch_news_sentiment(news_api_key)

print("\nNews Sentiment:")
for article in news_sentiment['articles'][:5]:
    print(article['title'])

# Fetch Twitter sentiment (replace 'your_twitter_api_keys' with your actual Twitter API keys)
twitter_api_key = 'your_twitter_api_key'
twitter_api_secret_key = 'your_twitter_api_secret_key'
twitter_access_token = 'your_twitter_access_token'
twitter_access_token_secret = 'your_twitter_access_token_secret'
twitter_sentiment = fetch_twitter_sentiment(twitter_api_key, twitter_api_secret_key, twitter_access_token, twitter_access_token_secret)

print("\nTwitter Sentiment:")
for tweet in twitter_sentiment[:5]:
    print(tweet.text)
```
